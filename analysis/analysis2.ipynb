```python
label = ['DE', 'JP', 'GB', 'US', 'HU', 'NZ', 'FR', 'IN', 'PK', 'CN',
        'GR', 'AE', 'NL', 'MX', 'CA', 'AT', 'NG', 'ES', 'PT', 'DK', 'IT',
        'HR', 'LU', 'PL', 'SG', 'RO', 'IQ', 'BR', 'BE', 'UA', 'IL', 'RU',
        'MT', 'CL', 'IR', 'CO', 'MD', 'KE', 'SI', 'CH', 'VN', 'AS', 'TR',
        'CZ', 'DZ', 'EE', 'MY', 'AU', 'IE']
```

```python
countData = []
count = 0

for item in label:
    if len(df[df['company_location']==item]) <= 5:
        count = count+1
    else:
        countData.append(len(df[df['company_location']==item]))
        
countData.append(count)
```

```python
percentage = []
for i in countData:
    percentage.append(i / sum(countData) * 100)
```

```python
percentlabel = ['DE', 'JP', 'GB', 'US', 'FR', 'IN', 'GR', 'CA', 'ES', 'Others']
```

```python
dfPercent = pd.DataFrame({"country":['DE', 'JP', 'GB', 'US', 'FR', 'IN', 'GR', 'CA', 'ES', 'Others'],
                         "percentage": percentage})

dfPercent = dfPercent.sort_values('percentage', ascending = False)
```

```python
plt.figure(figsize=(12,10))
colors = sns.color_palette('pastel')[0:5]
plt.pie(percentage, labels = percentlabel, colors = colors, autopct='%.0f%%')
plt.title("Distribution of data among different countries")
plt.savefig("../images/plot2_7.png" ,bbox_inches='tight')
plt.show()
```

##### 7. Plot 7: Distribution of data
This visualization displays the distribution of data categorized in each country, and the Others category consists of all countries that only consists of 5 or less data. It can be clearly seen that the number of data for United States is the highest among all the other countries. Compared to data in United States, data in all the other countries does not even count for half of the entire dataset. Thus, it is important to note that conclusions drawn from this dataset could be misunderstable due to the lack of number of data in other countries than United States.

## Method Chain to function.py

```python
def load_and_process(path, name):

    df1 = (
    pd.read_csv("../data/processed/ds_salaries_1.csv")
    .drop(columns=['work_year', 'employment_type', 'job_title', 'employee_residence', 'company_size', 'experience_level', 'salary_currency', 'salary'])
    .loc[lambda df: df["company_region"] == name]
    )
    df2 = (
          df1
          .assign(remote_ratio = lambda x : x['remote_ratio'].replace(0, "Not remote").replace(50, "Partially remote").replace(100, "Fully remote"))
          .groupby("remote_ratio")
          .mean()
          .reset_index()
      )

    return df2 

## function in codes file
```
